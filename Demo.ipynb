```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
from sklearn.linear_model import LinearRegression
data=pd.read_csv("temp_datalab_records_linkedin_company.csv")
```

```python
data
```

```python
x=data.groupby('industry').mean()
x.sort_values(by=['employees_on_platform'],ascending=False)
x1=np.array(x['employees_on_platform'])
x2=np.array(x['followers_count'])
plt.scatter(x1,x2)
plt.xlabel('Number of mean employees for each industry on the platform, log 10', fontdict=None, labelpad=None)
plt.ylabel('Number of mean followers count for each industry, log 10', fontdict=None, labelpad=None)

z = np.polyfit(x1,x2,1)
p = np.poly1d(z)
plt.plot(x1,p(x1),"r--")

```
